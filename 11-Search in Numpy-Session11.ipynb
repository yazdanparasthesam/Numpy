# numpy.where
numpy.where(condition, [ , x, y, ]/)

```python
##where mige boro begard donbale oon adadi ke tooye oon condition rooberoosh sedgh kone va rooye oon adad ke peyda mikoni bargharar bashe
##nokte inke ma mitoonim ham condition ro khali estefadeh konim va ham ba x va y va age condition ro khali estefadeh konim andis barmigardoone va age condition bargharar bood az arrayeye x entekhab
#kon va age condition bargharar nabood az arraeye y entekhab kon va bargardoon pas age condition True shod boro soraghe vector x va age False shod boro soraghe vector y
#va age ham khali bood andise moadele oon adadi ro bargardoon ke shart roosh bargharar boode

#dar mesale zir age arr>12 bashe miad va andis haye oon anaser ro barmigardoone
arr = np.array([10, 11, 12, 13, 14, 15])
np.where(arr>12)
```

```python
###hala mitoonam begam age in shart bargharar shode boro soraghe yek vector va age bargharar nashod boro soraghe yek vectore dg ,
#dar mesale zir farz mikonim ina adade feshare khoon hastan va age bozorgtar az 13 bood benevis high va dar gheyre in soorat ebenevis low
#pas bayad 2 ta vector baraye high va low taarif konam ke age sedgh kard ya nakard az har kodoom bardare ke tedade vector ha bayad ba tedade moghayese ha yeki bashe
arr = np.array([10, 11, 12, 13, 16, 18])
high = ["H", "H", "H", "H", "H", "H"]
low =  ["L", "L", "L", "L", "L", "L"]
np.where(arr>13, high, low) #inja ham behesh migam ke age shart True bood yani bozorgtar az 13 shod az high bardare va age bozorgtar az 13 nashod az low bardare
```

```python
##hala haminja mitoonam tooye shart ebarate manteghi mesle AND va OR dashte basham va migam masalan age 24 bood be jaye high benevis outband chon dg 24 baraye feshare khoon gheyre manteghie
arr = np.array([10, 11, 12, 13, 16, 24])
high = ["H", "H", "H", "H", "H", "Out-band"]
low =  ["L", "L", "L", "L", "L", "L"]
np.where((arr>13)& (arr>=24), high, low)
```

```python
##hamin ro yek arrayeye 2 bodi ham mitoonim benevisim ke age sharte cond bargharar beshe miad soraghe in matris va age bargharar nashe mire soraghe oon yeki matris
#np.where(con, [[],[]], [[],[]])
#mesalesh mesle zir mishe ke farz mikonim too ghesmate condition yeki rafte va natayeje amadeye True/Flase ovorde
np.where([[True, False],[True, True]], [[1,2],[3,4]], [[5,6],[7,8]])
##chon onsore avvale condition matrix True hast , 1 ro barmidare va onsore badi False hast pas 6 ro bar midare va 2 ta onsore baadi ham True hast pas 3 va 4 ro bar midare
#pas farghesh ba 1 bodi ine ke be jaye vector ya bordar inja matris daram che baraye dorost boodane shart va che nadorost boodane shart
```

```python
#extract condition va array ro migire va khode oon adad haro bar migardoone yani age oon condition roosh sedgh kard khode adad ro bar migardoone
#aval yek data mesle zir dorost mikonam
arr = np.arange(10).reshape(2,5)
arr
```

```python
#hala kole adade in arraye ro  -4 kam mikonam
arr -4
```

```python
#va hala migam age inkaro kardam va >= az 0 bood bargardoon pas condition e man intori mishe
arr -4 >= 0
```

```python
#hala in shart ro be extract midam
print(arr -4)
np.extract(arr -4 >= 0,arr)
```

```python
##ya mitoonam yek condition e dg begzaram va begam ke kole adade arraye ro bar 4 taghsiim kon va faghat  oonayi ke baghimandashoon mokhalefe 0 hast ro neshoon bede
print(arr)
np.extract(np.mod(arr, 4)!=0,arr)
```
