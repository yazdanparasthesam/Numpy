```python
#to work with excel in python we have to import xlrd
import numpy as np
from sys import getsizeof as size
import io
import copy
import xlrd #should be installed with pip install
```

```python
#yek variable wb ya workbook dorst mikonam va behesh path file excel ro midam
wb = "/content/sample_data/Baseball salaries.xls"

#hala yek function taarif mikonam ke data haye in excel ro bekhoonam va be numpy array tabdil konam ke wb ro be onvane voroodi migire
def create_array(wb):
    sheet = xlrd.open_workbook(wb).sheet_by_index(0) #ba 'xlrd.open_workbook' wb ro baz karde va ba 'sheet_by_index' miam va sheet 0 ro mikhoonam
    #pas alan sheet0 tooye variable sheet gharar gerefte
    #hala mikham mohtaviate sheet ro line by line bekhoonam chetor???
    wb_list = [sheet.row_values(i) for i in range(sheet.nrows)] ##miam ba row.values khat be khate file excel ro mikhoonam va tooye yek list migzaram
    return np.array(wb_list) #hala in list ro be yek numpy array tabdil mikonam

create_array(wb)
#hamintor ke dar zir mibinid har satr az excel ro tabdil be list kard
```

```python
##hala in ro mirizam tooye yek variable be esme 'arr' va bahash mitoonam kar konam
arr = create_array(wb)
#header jadval be darde mohasebate man nemikhore va hala farz konim satre akhar ham be darde ma nemikhore  intori hazfeshoon mikonam :
#arrayeye 'arr' ro slice mikonam va mirizam all_data
#ke chetori slice mikonam ?? migam az satre aval be baad yani satr avval ro neshoon nade va satre akhar ham nemikham , va age bekham arraye ro az payin bekhoonam
#satre akhare mishe -1 va bad -2 va bad -3 va ... ke slice mikonam az 1 ta -1 yani satre avval va akhar ro nemikham
all_data = arr[1:-1]
all_data
```

```python
##age too bala deghat konid hata adad ha ham string hastan chon datatype taarif nakardim va age tooye return function
#datatype ro object begzarim adad haro be soorate float neshoon mide
def create_array1(wb):
    sheet = xlrd.open_workbook(wb).sheet_by_index(0)
    wb_list = [sheet.row_values(i) for i in range(sheet.nrows)]
    return np.array(wb_list, dtype = object)

create_array1(wb)
arr = create_array1(wb)
all_data = arr[1:-1]
all_data
```

```python
##hala age bekham be har column az all_data access dashte basham ke mitoonim yek seri constant taarif konim va column name haro behesh bedim va
#ba in column name ha data ro begirim ke in programming ro khanatar mikone
TEAM = 0
PLAYER = 1
SALARY = 2
POSITION = 3
salary_data = all_data[:, SALARY].astype(float) ##in mige data ro hameye satr hash ro ba sotoone SALARY bardar biar
salary_data
```

```python
#hala yek function e be esme stat_calcs minevisam baraye mohasebeye parameter haye amari ke voroodi hash hamin salary_data va all_data hast
#ke salary_data esmesh sdata migzaram va all_data esmesh data hast

#https://numpy.org/doc/stable/reference/routines.statistics.html

def stat_calcs(sdata, data):
  print("Average of all salaries is \t ", sdata.mean().round(2)) ##mitoonestam intori ham hesab konam np.mean(sdata).round(2)
  print("Median of all salaries is \t ", np.median(sdata).round(2))
  print("Maximum of all salaries is \t ", np.amax(sdata).round(2))
  print("Minimum of all salaries is \t ", np.amin(sdata).round(2)) ###nanmax va nanmin baraye mavaredi ke bekham nan haro ignor kone
  print("SD of all salaries is \t ", np.std(sdata).round(2))
  print()

  ind_max = sdata.argmax() ##in behem indexe maximum ro mide ke ba in index mitoonam baghye mavaredesh ham begam yani begam masalan bishtarin salary baraye
  #felan bazikon hast ke position esh felan hast va too felan team hast
  print("Maximum Salary is %d which earned by %s "\
        "who are playing in %s team in %s position"%(sdata[ind_max], data[ind_max,PLAYER], data[ind_max,TEAM], data[ind_max,POSITION]) ##%d ro az sdata gereftam va
        #baghyeye mavared mesle team , playername
        # va ... ro az data migiram chon shamele all_data hast
        )

  ####hala hamin ro baraye minimum ham mitoonam hesab konam
  ind_min = sdata.argmin() ##in behem indexe minimum ro mide
  print("Minimum Salary is %d which earned by %s "\
        "who are playing in %s team in %s position"%(sdata[ind_min], data[ind_min,PLAYER], data[ind_min,TEAM], data[ind_min,POSITION]) ##%d ro az sdata gereftam va
        #baghyeye mavared mesle team , playername
        # va ... ro az data migiram chon shamele all_data hast
        )

###hala az function e ke neveshtam miam biroon va behesh voroodiharo midam
print("******All Calculations******")
stat_calcs(salary_data, all_data)

```

```python
#midamesh be set ke tekrari haro hazf kone va arrayey yekta beshe chon set majmooe mikone(az in alamat {} mishe fahmid) va tekrari haro hazf mikone
set(all_data[:,TEAM])
```

```python
###hala man mitoonam yek function taarif konam ke mohasebat ro baraye har team joda anjam bede
def calcs_for_each_team():
  for team_name in set(all_data[:,TEAM]): #migam yek for rooye in majmooe bezan be ezaye har team
    mask = (all_data[:,TEAM] == team_name) #halaja mask ro taarif mikonam ke miad all_data ro mask ya filter mikone vaghti esme sotoon har dafe yek chiz hast
    #masalan age esme team ARI  bood miad va all_data ro rooye column e team mask mikone vaghti esme team ARI hast
    print()#baraye andakhtane fasele
    print("******For %s Team******"%team_name)
    stat_calcs(salary_data[mask], all_data[mask])

calcs_for_each_team()
```
