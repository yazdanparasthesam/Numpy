```python
#example for badformat data
#farz konid man data6 ro be in shekl daram ke yek seri % ham dare va vaghti ino be genfromtxt midam oon adad haye % dar ro baram nan neshoon mide chon mige man character e % ro nemifahmam tooye
#float chon defaukt man float hast
data6 = "1, 2.3%, 56.\n 7, 89.99%, 0"
np.genfromtxt(io.StringIO(data6), delimiter=",")
#pas bayad biam correction bezanam va % ro mafhoome riazish ro piadeh konam masalan 2.3% mishe 2.3/100
```

```python
##pas b a lambda yek function minevisam ke alamate % ro strip mikone va baad taghsim be 100 mikone
data6 = "1, 2.3%, 56.\n 7, 89.99%, 0"
convertor = lambda x :float(x.strip("%"))/100.
np.genfromtxt(io.StringIO(data6), delimiter=",", converters={1:convertor})  ##hala miam va in converters ro rooye sotoone avval emal mikonam
```

```python
##faghat oomade ba namade elmi neshoonesh dadeh pas baraye dorost kardanesh az names estefadeh mikonim
data6 = "1, 2.3%, 56.\n 7, 89.99%, 0"
convertor = lambda x :float(x.strip("%"))/100.
np.genfromtxt(io.StringIO(data6), delimiter=",", converters={1:convertor}, names = "A, B, C")

#in badformat data ro ba excel va notepad++ mishe dorostesh kard chon notepad++ regex dare
```

```python
#example for missing data
#yek convertor minevisam ke chrarctere khali ro strip kon(vaghti too strip hichi nagzaram yani charactere khali ro strip kon) ya be jash manfie 1000 begzar chon age
#jaye khali bashe error migire va mige jaye khali ba formate float nemikhoone va adad mikham
data7 = "1, ,3\n5, 6, 7"
convertor = lambda x :float(x.strip() or -1000)
np.genfromtxt(io.StringIO(data7), delimiter=",", converters={1:convertor}, names = "A, B, C")
```

```python
###hala negah konim age intori benevisim chi mishe ???
data7 = "1, ,3\n5, 6, 7"
convertor = lambda x :float(-1000)
np.genfromtxt(io.StringIO(data7), delimiter=",", converters={1:convertor}, names = "A, B, C")

##mibinim ke jaye sotoone avval kolan -1000 gozasht ke khoob nist yani 6 ro ham ba -1000 replace kard pas x.strip kari ke too code mikone ine ke space ghable 6 ro mizane va mibine ke adad dare pas ba
#-1000 replace nemikone pas ya strip mikone va age adad bashe chizi bar nemigardoone ya age kolan space bashe -1000 jash migzare
```

```python
#example for missing data
data8 = "N/A, 2, 3\n4, ,???" ###N/A yani not available va jaye khali va ??? har 3 ta missing data ma hastan
kwargs = dict(delimiter=",",
               dtype=int,
               names="A, B, C",
               missing_values={0:"N/A", "B":" ", 2:"???" }, ###inja goftam ke too sotoone aval "N/A" hast tooye sotoone dovom space ya fazaye khali hast ke hala man ham mitoonam be jaye sotoone
               #dovvom 1 begzaram va begam charactere space yani : 1:" " ham mitoonam esme sotoone dovom ro bedam : "B":" " yani esme sotoon ro ke bala taarif kardim gharar bedam
               #va tooye sotoone sevom ham ke charactere missed shodeye man ??? hast
               filling_values={0: 10,'B': 9, 2:-100 }  ###hala inja migam ke oon missing values ro ba chi por kone
               )
np.genfromtxt(io.StringIO(data8), **kwargs) ###inja be jaye inke tak tak parameter haro gharar bedam miam va kwargs ro taarif mikonam va kwargs ro be onvane yek dictionary taarif mikonam
```

```python
#man mitoonam yek tuple az mfilling values taarif konam va begam boro ino tooye jahaye khalie data insert kon
####example for missing data
data9 = """
1,2,,,5
6,,7,,
10,,,,
"""
filling_values = (100, 200, 300, 400, 500) ##intori vaghti taarif mikonam yani boro too sotoone avval begard va harja khali bood 100 begzar va too sotoone dovom begard har ja khali bood 200 bezar va ...
np.genfromtxt(io.StringIO(data9), delimiter=",", filling_values=filling_values)
```
