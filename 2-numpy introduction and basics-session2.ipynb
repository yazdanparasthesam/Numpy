```python
import time
import numpy as np

size_vector = 1000000

def list_view():
  t1 = time.time() ###time alan ro dar nazar migire
  l1 = range(size_vector) #yek list taarif mikonam ke baram az 1 ta 100000 yek vector tolid mikone
  l2 = range(size_vector)
  summation = [l1[i] + l2[i] for i in range(len(l1))] ##Mikham in 2 ta list ro ba ham jam konam
  return time.time() - t1 #dobare time alan ro return kon va azash t1 ro kam kon ke yani cheghadr tool keshide in summation anjam beshe

def numpy_view():
  t1 = time.time()
  n1 = np.arange(size_vector)
  n2 = np.arange(size_vector)
  summation = n1 + n2  ##inja numpy be ma yek ghdrate khoobi mide va mitoone jame vector hash ro ba hamin alamte + , behem bede va in dar haghighat hamoon kare bala ro anjam mide vali chon numpy hast
  #kheyli sadeh tar pas numpy dar vector base amal mikone vali list inkaro nemitoone bokone va error migire
  return time.time() - t1


t_list_view = list_view()
t_numpy_view = numpy_view()

print("Numpy is" + str(t_list_view/t_numpy_view) + "is faster than list ")

###mibinim ke numpy dare in ro 24 barabar sari tar anjam mide ke in kheyli zamane ghabele tavajohi hast
```
