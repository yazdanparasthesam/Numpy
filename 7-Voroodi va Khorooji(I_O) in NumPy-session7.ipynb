```python
#yek arraye be shekle zir taarif mikonam
import numpy as np
arr = np.array([i+j for i in range(2) for j in range(2)])
print("Array is:", arr)
```

```python
#ke albate age too code bala jaye [] ro avaz konim object esh ro behem neshoon mide ke ba in kari nadarim va ma rooye balayi kar mikonim
arr1 = np.array([i+j] for i in range(2) for j in range(2))
print("Array is:", arr1)
```

```python
#hala age bekham ino zakhire konam  ba np.save() inkaro mikonam
#vali chon formattesh binary hast nemitoonam bebinam , bayad loadesh konam ya formattesh ro be UTF8 ya ASCII change konam ke btoonam bebinam
np.save("hesamarray", arr) ##inja esme file ro ke bedam mire tooye current working path zakhire mikone ke mitoonam behesh path ham bedam
print("hesamarray.npy")
```

```python
print(np.load("hesamarray.npy"))
```

```python
##save kardan ba formate npz va load kardan e mojadade oon
np.savez("hesamarrayz", arr) ##using *args amalan too in halat daram az raveshe *args estefadeh mikonam
loaded_array = np.load("hesamarrayz.npz")
print(loaded_array)  ###ke mibinim dare object mano barmigardoone na khode array ro va elatesh ham in hast ke too formate npz content ya mohatavaye arraye dakhele
#yek file ya container zakhire mishe ke esmeshro behehm dadeh dar zir "arr_0" pas mohatavaye arraye rafte tooye file "arr_0" dar objecte in npz
```

```python
##pas mohtavaye oon file ro intori mibinam
print(loaded_array['arr_0'])
```

```python
np.savez("hesamarrayz", x = arr) ##using **kwds amalan too in halat daram az raveshe **kwds estefadeh mikonam ke masalan inja be jaye inke khode arraye ro
#bedam , barash yek name mesle 'x' dar nazar migiram
loaded_array = np.load("hesamarrayz.npz")
print(loaded_array)
print(loaded_array['x']) ###va inja in esm ro be jaye 'arr_0' seda mizanam ke chon dar halate default oon 'arr_0' ro be onvane name e  file haye dakhelie arraye
# dar nazar nemigire va be jash hamoon 'x' ro seda mizanam

#pas vaghti x = arr ro migzaram in yek halate keyword ya **kwds hast ke in x esme arrayeye man dar npz hast ke mitoonam ba eshare be andise x behesh dastresi dashte
#basham
```

```python
##dar session ghabli np.loadtxt ro barresi kardim hala mikham bebinam ke age bekham savetxt estefadeh konam che option hayi daram
#avval yek arraye be esme 'x' misazam ba np.arange ke 20 ta ozv dashte bashe va reshape esh ham mikonam ke 4 ta satr dashte bashe va 5 ta sotoon
x = np.arange(20).reshape(4, 5)
x
```

```python
#hala age ino bekham ba savetxt , save esh konam , avval bayad behesh yek name(fname) bedam ke masalan migzaram 'x_array' va bad behesh esme arraye e ke mikham zakhire
#konam bedam ke inja esmesh 'x' hast va bad bayad behesh format ya fmt bedam ke ba format taarif mishe ke chetori oon ro tooye file neshoon bede
#ke formate default esh '%.18e' hast ke mige bakhshe sahih ke hichi vali bakhshe asharie oon 18 ta raghame ashari begire va chon 'e' hast yani mishe namade elmi
#delimiter ke mige joda konandeh am chi bashe ke dar halate default ' ' ya space hast
#hala age code zir ro ba mavarede bala ke goftim ejra konim file 'x_array' ro baramoon ijad mikone
x = np.arange(20).reshape(4, 5) #1D and 2D tavajoh konid ke savetext baraye arraye haye 1 bodi va 2 bodi estefadeh mishe va baraye arraye haye 3 bodi va bishtar
# ghabele estefadeh nist
np.savetxt('x_array.txt', x, delimiter=' ')

#ke too file ijad shode mibinam ke hameye adad ha ba 18 raghame ashar hast va delimiter space hast
```

```python
#mitoonam delimiter ro ',' begzaram
np.savetxt('x_array.txt', x, delimiter=',')
```

```python
#mitoonam format ro '%.2f' begzaram yani bade bakhshe sahih ta 2 raghame ashari bashe va 'f' dar inja yani 'float' bahse
np.savetxt('x_array.txt', x, delimiter=',', fmt='%.2f')
```

```python
#ya mitoone ta 3 raghame ashari bashe
np.savetxt('x_array.txt', x, delimiter=',', fmt='%.3f')
```

```python
#ya hata mitoonim be soorate integer neshoon bedim
np.savetxt('x_array.txt', x, delimiter=',', fmt='%d')
```

```python
# ya hata mitoonam '%3d' begzaram ke baraye har adad 3 ta position lahaz mikone va age maximum adad hamoon 2 raghami bashe '2d' va
#age maximum adad hamoon 3 raghami bashe '3d' migzarim
np.savetxt('x_array.txt', x, delimiter=',', fmt='%3d')
#inja  chon maximum adad hamoon 2 raghami hast az hamoon '2d' estefadeh mikonam
```

```python
# ke mitoonam behesh begam baraye adade yek raghami space haye khali ro ba 0 por kone
np.savetxt('x_array.txt', x, delimiter=',', fmt='%02d')
```

```python
#in kar ro baraye ashari ha ham mitoonam bokonam
np.savetxt('x_array.txt', x, delimiter=',', fmt='%2.2f')
```

```python
#hala baraye inke adad ha zire ham gharar begiran migam ke bozorgtarin adad yani '19.00' chand ta position dare?
#5 ta
#pas minevisam  %5.2f
np.savetxt('x_array.txt', x, delimiter=',', fmt='%5.2f')
#option e 'comments' ro dar 'genfromtxt' migam
```

```python
#numpy.fromregex
file = open('my_reg.txt', 'w')## yek file be esme 'my_reg' dorost mikonam va mikham format esh 'w' bashe yani dakhelesh
# benevisam
_ = file.write("1234 hesam\n5678 Alireza\n91011 uvwxyz") #ba file.write inha ro tooye file neveshtam va badesh be variable
#ba esme _ assign kardam ke variable ba esme '_' yani mirizam file ro tooye yek variable vali baram mohem nist
# ke esme variable chi bashe
file.close() #file ro mibandam
#age code bala ro ejra konam baram file 'my_reg' ro misazeh

#hala regex am ro taarif mikonam
regexp = r"(\d+)\s+(.....)"##in mige ke avval adad haro joda kon (ba \d) va badesh space dashte bashim (ba \s) va badesh 5 ta character dashte bashim ba (.....)
#nokteye mohem injast ke in paranthes ha baese tashkile groupi az oonha mishe yani alan integer ha mishan yek group va character ha mishan yek group e dg pas man
#alan 2 ta category daram adad va string
out = np.fromregex('my_reg.txt', regexp, dtype=[('num', np.int64), ('key', 'S5')]) ##inja miam va migam voroodim bashe file my_reg.txt va regexp ham ke bala taarif kardam ro be in file emal kon
#va data type ham 2 model bood dg adad va string ke yek tuple taarif mikonam va baraye adad esme 'num' ro taarif mikonam ba type 'int64' va baraye string ham esme 'key'
#ro taarif mikonam ba type 'S5'
out  ###mibinim behemoon formate arraye bargardoond yani out[0] ma mishe ( 1234, b'hesam'
```
