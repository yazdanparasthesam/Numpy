```python
c[0] = 9
print(c, z)
```

# *Frombuffer*

```python
st = b'hesam'
np.frombuffer(st, dtype='S1', count = -1)
##count moshakhas mikone ke chand ta az oon buffer ro bekhoonam va age count barabare -1 bashe yani hameye oon ro bekhoon
##dar zir mibinim ke chizi ke bargardoone ine ke oomade kole string ke binary format hast ro khoonde va 1-dimensional too yek arraye gharar dadeh
```

```python
##yek option offset ham daram ke mige az koja shoroo kon be khoondan masalan age begam offset 1 va count 2 mire az 'e' tooye hesam shoroo mikone va 2 ta mikhoone
np.frombuffer(st, dtype='S1', count = 2, offset = 1)
```

```python
##offset dar halate default 0 hast yani age moshakhas nashe az avval shoroo mikone be khoondan
np.frombuffer(st, dtype='S1', count = 2)
```

```python
np.frombuffer(st, dtype='S1', count = 1, offset = 3) ##in mige az offset e 3 shoroo kon va 1 character bekhoon
```
