```python
###COVARIANCE noe ertebate 2 ta variable ro bayan mikard ke migoft ertebat positive hast ya negative hast ya aslan  ertebati ba ham nadaran
###correlation coefficients migoft ke covariance dare type hambastegi ro moshakhas mikone vali metrici behetoon nemide ke cheghdar ina be ham vabaste hastan yani age begam in 2 moteghayer beham
#positive hastan , covariance nemitoone behem bege ke cheghadr in 2 moteghayer az nazare type beham positive hastan , covariance nemitoone bege in 2 moteghayer cheghadr beham vabastegi daran
#vali correlation coefficients in shedate vabastegi ro bayan mikone be ebarate dg covariance too dele correlation penhan hast
```

```python
#Explain Covariance and Correlation :

#avval 2 ta seri adad dar ghalebe bordar taarif mikonam ke avvali x ke yek seri adade randon integer behem bede va dameye inha beyne 0 ta 50 bashe va 1000 ta adad baram tolid kone:
import numpy as np
x = np.random.randint(0, 50, 1000)

#y ro ham mesle bala taarif mikonam ba in tafavot ke in adade random ro be ezafeye hamoon x , ke taarif kardam mikonam ke rabeteye positive in 2 ro neshoon bedam

y = x + np.random.normal(0, 50, 1000)##ke in mige az random estefadeh kon vali az tozie amarie normal estefadeh kon va dameye inha beyne 0 ta 50 bashe va 1000 ta adad baram tolid kone
#mamoolan vaghti mikham yek noise be dadeh ha ezafe konam az tozie normal estefadeh mikonam pas mamoolan noise ro az adadi estefadeh mikonam ke random hastan vali tozie normal daran
```

```python
##ba function e corrcoeff mitoonam barresi konam ke x va y che noe vabastegi baham daran
np.corrcoef(x, y)
#anvae correlation ha vojood dare vali in formul az hamoon pearson estefadeh mikone
#behem yek matris mide ke dar in matris satr haye man hamon x,y hastan va sotoon ha ham hamoon x,y hastan
#pas ertebate x,x mishe 1 ke yani 100% ertebate positive, ertebate x,y mishe 0.31 ke yani 31% ertebate positive ,ertebate y,x mishe 0.31 va ertebate y,y mishe 1 pas +0.31 ke alamate + mishe sign esh
# hast ke bayan konandeye noe ertebat hast va bozorgie adade 0.31 ham behesh migan magnitude ke bayan konadeye shedate ertebat hast
#nokte: inke in matris motegharen hast yani x ba y har ertebati dashte bashe , y ba x ham hamoon meghdar bahash ertebat dare
#nokte: har che adade corrcoef be 1 nazdik tar bashe yani ina bishtar vabaste and va har che az 1 doortar bashe yani kamtar beham vabaste and.
```

```python
###hala age bekham nemoodare in ertebat ro bekesham ke betoonam in ertebat ro bebinam az matplotlib estefadeh mikonam

###age bekham ertebate x ba y ro neshoon bedam az che nemoodari estefadeh konam ???? scattter plot ke az hasele zarbe cartesian be dast miad

#ma too riazi chizi be esme haselzarbe cartesian darim ke 2 ta majmooe ro baham zarbe cartesian mikone mesle zir:
#(0, 1, 2) cartesian(1,2,3) ---> (0,1)(0,2)(0,3)(1,1)(1,2)(1,3)(2,1)(2,2)(2,3)
print(np.corrcoef(x, y)) ##26% ertebat daran
import matplotlib.pyplot as plt
plt.scatter(x, y)
plt.xlabel("X")
plt.ylabel("Y")
plt.legend(loc= 'upper left') ##taarife legend bala samte chap
```

```python
##mitoonim az baghyeye option haye matplotlib estefade konam  ke "c" baraye range hast ke chon inja "g"dadim rang ro sabz mikone va "label" ke ba label miam va labele legend ro tayin mikonam
#ba marker shekle barg ro be jaye noghat dadam va ba alpha neshoon midam range har noghte cheghadr kamrang ya porrang bashe
plt.scatter(x, y, c="g", alpha=0.5, marker=r'$\clubsuit$', label="Corr")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend(loc= 'upper left')
##binim ke dare mire be samte yek khate saf ke age khate saf bashe yani correlation e baham nadaran
```

```python
###miam dameneye y ro be jaye 0 ta 50 be 0 ta 5taghyir midam va correlation intori mishe
x = np.random.randint(0, 50, 1000)
y = x + np.random.normal(0, 5, 1000)
print(np.corrcoef(x, y)) #mibinim ertebat 94% shode
plt.scatter(x, y, c="g", alpha=0.5, label="Positive Corr")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend(loc= 'upper left')
plt.show() #vaghti plt.show mikonam age dar edame ham az inha estefadeh kardam in graphesh mimoone
```

```python
##hala baraye negative correlation ham yek mesal bezanim va hamoon balayi hast faghat y ro az 100 kam mikonam
k = np.random.randint(0, 50, 1000)
t = 100 - k + np.random.normal(0, 5, 1000)
print(np.corrcoef(k, t)) #mibinim ertebat -94% shode
plt.scatter(k, t, c="r", alpha=0.5, label="Negative Corr")   ##indafe rang ro "r" gozashtam yani red
plt.xlabel("K")
plt.ylabel("T")
plt.legend(loc= 'upper right')
plt.show()
```

```python
###hala baraye halati ke aslan ertebati nadaran
r = np.random.randint(0, 50, 1000)
u = np.random.randint(0, 50, 1000)
print(np.corrcoef(r, u)) ##ertebate kheyli kam yani 2%
plt.scatter(r, u, c="b", alpha=0.5, label="Zero Corr")   ##indafe rang ro "b" gozashtam yani blue
plt.xlabel("R")
plt.ylabel("U")
plt.legend(loc= 'upper left')
plt.show()
```

```python
##Correlation Matrix : hamoon data haye bala ro miaram va corrcoef hame ro too yek matris migiram
x = np.random.randint(0, 50, 1000)
y = x + np.random.normal(0, 5, 1000)
t = 100 - x + np.random.normal(0, 5, 1000)
r = np.random.randint(0, 50, 1000)

data  = np.corrcoef([x, y, t, r])
data
#in mishe correlation matrix e man be soorate adadi

```

```python
#baraye plot kardane ertebate inha az heatmap estefadeh mikonam

##az imshow va cmap ke hamoon colormap hast estefade mikonam ke colormap dar haghighat yek teife range khas hast
plt.imshow(data, cmap = "cool")

```

```python
##chon kheyli khoob neshoon nemide miam az xticks va yticks estefadeh mikonam ke ba xticks va yticks miam va mehvare x ha va  mehvare y ha ro be ticke haye mokhtalef taghsim mikonam
#va too har kodoom az in tike ha yeki az dataset haro migzaram ke mishe hamoon x,y, t, r
plt.imshow(data, cmap = "cool")
plt.xticks(np.arange(4), ('x', 'y', 't', 'r'))
plt.yticks(np.arange(4), ('x', 'y', 't', 'r'))

#ke mibinim khorooji be shekle zir mishe
```

```python
###hala yek colorbar taarif mikonam ke  kamtarin va bishtarin meghdare colorbar ba vmin va vmax tayin mishe
plt.imshow(data, cmap = "cool", vmin=-1, vmax=1)
plt.colorbar()
plt.xticks(np.arange(4), ('x', 'y', 't', 'r'))
plt.yticks(np.arange(4), ('x', 'y', 't', 'r'))
plt.show()
```
